```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('james_bond_data.csv')
data.head(10)
print(data)
```

```python
plt.scatter(data["Avg_User_IMDB"], data["Avg_User_Rtn_Tom"])
plt.xlabel("IMDB Ratings")
plt.ylabel("Rotten Tomatoes Ratings")
plt.title("IMDB Rating vs Rotten Tomatoes Rating")
plt.show()
correlation = data["Avg_User_IMDB"].corr(data["Avg_User_Rtn_Tom"])
print(correlation)
# Yes, there seems to be a strong correlation between IMDB rating and Rotton Tomatoes ratings
```

```python
plt.scatter(data["Avg_User_IMDB"], data["Kills_Bond"])
plt.xlabel("IMDB Ratings")
plt.ylabel("Bad Guys Killed")
plt.title("Bond Kills vs IMDB Ratings")
plt.show()
correlation = data["Avg_User_IMDB"].corr(data["Kills_Bond"])
print(correlation)
# No, there does not seem to be any correlation between IMDB rating and number of bad guys killed by Bond
```
